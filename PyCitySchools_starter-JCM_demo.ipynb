```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data_complete.head()
```

```python
# calculate total number of schools using .unique and len
schools_count = len(school_data_complete["school_name"].unique())
print(f"The total number of schools is: {schools_count}")
```

```python
# calculate total number of students using .count
student_count = school_data_complete["Student ID"].count()
print(f"The total number of students is: {student_count}")
```

```python
# calculate total budget
total_budget = school_data["budget"].sum()
print(f"The total budget for schools is: {total_budget}")
```

```python
# calculate avergage math scores
avg_math_scores = school_data_complete["math_score"].mean()
print(f"The average math score is: {avg_math_scores}")
```

```python
# calculate avergage reading scores
avg_reading_score = school_data_complete["reading_score"].mean()
print(f"The average reading score is: {avg_reading_score}")
```

```python
# calculate count passing math
count_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
print(f"The count of students passing math is: {count_passing_math}")
```

```python
# calculate percentage passing math
percentage_passing_math = count_passing_math / student_count * 100
print(f"The percentage of students passing math is: {percentage_passing_math}")
```

```python
# calculate count passing reading
count_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
print(f"The count of students passing reading is: {count_passing_reading}")
```

```python
# calculate percentage passing reading
percentage_passing_reading = count_passing_reading /  student_count * 100
print(f"The percentage of students passing reading is: {percentage_passing_reading}")
```

```python
# calculate count of students passing math and reading
count_passing_math_and_reading = school_data_complete[(school_data_complete["math_score"] >= 70)
                                                     & (school_data_complete["reading_score"] >= 70)].count()["student_name"]
print(f"The count of students passing both math and reading is: {count_passing_math_and_reading}")
```

```python
# calculate overall passing math and reading rate
overall_passing_reading_math_rate = count_passing_math_and_reading / student_count * 100
print(f"The percentage of students passing both math and reading is: {overall_passing_reading_math_rate}")
```

```python
# Create a District Summary Table (a DataFrame) using pd.DataFrame
District_Summary = pd.DataFrame({
    "Total Schools": [schools_count],
    "Total Students": [student_count],
    "Total School Budget": [total_budget],
    "Average Math Score": [avg_math_scores],
    "Average Reading Score": [avg_reading_score],
    "% Passing Math": [percentage_passing_math],
    "% Passing Reading": [percentage_passing_reading],
    "% Overall Passing": [overall_passing_reading_math_rate]
})
District_Summary
```

```python
# Format District Summary (DataFrame)

District_Summary["Total Students"] = District_Summary["Total Students"].map("{:,}".format)
District_Summary["Total School Budget"] = District_Summary["Total School Budget"].map("${:,.2f}".format)
District_Summary = District_Summary.round(2)
District_Summary
```

```python
school_data
```

```python
# Determine School Type
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
# calculate total_students_per_school using value_counts
total_students_per_school = school_data_complete["school_name"].value_counts()
total_students_per_school
```

```python
school_data_complete.head()
```

```python
# calculate per school budget
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_school_budget
```

```python
# calculate per school spending per student (i.e. per_school_capita)
per_school_capita = per_school_budget / total_students_per_school
per_school_capita
```

```python
# calcualte average math score per school
avg_per_school_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
avg_per_school_math
```

```python
# calculate average reading score per school
avg_per_school_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
avg_per_school_reading
```

```python
# create a DataFrame of students who passed math
school_students_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]
school_students_passing_math
```

```python
# calculate % passing math
per_school_passing_math = school_students_passing_math.groupby(["school_name"]).count()["student_name"] / total_students_per_school * 100
per_school_passing_math.round(2)
```

```python
# create a DataFrame of students who passed reading
school_students_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]
school_students_passing_reading
```

```python
# calculate the percentage of students passing reading
per_school_passing_reading = school_students_passing_reading.groupby(["school_name"]).count()["student_name"]/ total_students_per_school * 100
per_school_passing_reading
```

```python
# create a DataFrame of students who passed math and reading
school_students_passing_math_and_reading = school_data_complete[(school_data_complete["math_score"] >= 70) 
                                                               & (school_data_complete["reading_score"] >= 70)]
school_students_passing_math_and_reading
```

```python
# calculate the percentage of students passing math and reading
per_school_passing_math_and_reading = school_students_passing_math_and_reading.groupby(["school_name"]).count()["student_name"]/total_students_per_school * 100
per_school_passing_math_and_reading
```

```python
# Create a dataframe to hold the above results
School_Summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": total_students_per_school,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Math Score": avg_per_school_math,
    "Average Reading Score": avg_per_school_reading,
    "% Passing Math": per_school_passing_math,
    "% Passing Reading": per_school_passing_reading,
    "% Overall Passing": per_school_passing_math_and_reading
})
School_Summary
```

```python
# Format School Summary
School_Summary = School_Summary.round(2)
School_Summary["Total Students"] = School_Summary["Total Students"].map("{:,}".format)
School_Summary["Total School Budget"] = School_Summary["Total School Budget"].map("${:,.2f}".format)
School_Summary["Per Student Budget"] = School_Summary["Per Student Budget"].map("${:,.2f}".format)

School_Summary
```

```python
# create a data series of scores using conditions for each grade 9 - 12
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]
```

```python
# Use groupby to find the average math scores by grade
ninth_graders_math_scores = ninth_graders.groupby(["school_name"]).mean()["math_score"]
tenth_graders_math_scores = tenth_graders.groupby(["school_name"]).mean()["math_score"]
eleventh_graders_math_scores = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
twelfth_graders_math_scores = twelfth_graders.groupby(["school_name"]).mean()["math_score"]
```

```python
# create a DataFrame to hold the series
Math_scores_by_grade = pd.DataFrame({
    "9th": ninth_graders_math_scores,
    "10th": tenth_graders_math_scores,
    "11th": eleventh_graders_math_scores,
    "12th": twelfth_graders_math_scores
})
Math_scores_by_grade.round(2)
```
